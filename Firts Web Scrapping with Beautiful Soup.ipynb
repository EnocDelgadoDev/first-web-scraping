# Import libraries

```python
from bs4 import BeautifulSoup
import requests
```

# Extract HTML code

```python
web_site = 'https://subslikescript.com/movie/Avengers_Endgame-4154796'
result = requests.get(web_site)
# Get text of this variable
content = result.text
```

```python
soup = BeautifulSoup(content, 'lxml')
# prettify allows us to print our content in the best way possible
print(soup.prettify())
```

# Get data

```python
# Get main element
# find is a method that allows us to obtain specific data, it has two arguments(tag, attributes)
box = soup.find('article', class_='main-article')

# Get an element without attribute
title = soup.find('h1').get_text()

# Get transcript
# strip removes blanks before and after a string
# separator replaces a line break
transcript = soup.find('div', class_='full-script').get_text(strip=True, separator=' ')


# print
print(title)
print(transcript)
```

# Export data

```python
# use the open method to export our data into text
with open(f'{title}.txt', 'w') as file:
    # write inside the parenthesis our data
    file.write(transcript)
```
